```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dominus = [1, 2, 3, 5]
functus = [3, 4, 2.5, 0.5]
```

```python
aa = sum([x**4 for x in dominus])
ab = sum([x**3 for x in dominus])
ac = sum([x**2 for x in dominus])

ba = sum([x**3 for x in dominus])
bb = sum([x**2 for x in dominus])
bc = sum([x for x in dominus])

ca = sum([x**2 for x in dominus])
cb = sum([x for x in dominus])
cc = 1

M = np.array([[aa, ab, ac], 
            [ba, bb, bc], 
            [ca, cb, cc]])
```

```python
z1 = sum([x**2 * y for x, y in zip(dominus, functus)])
z2 = sum([x * y for x, y in zip(dominus, functus)])
z3 = sum([y for y in functus])

z = np.array([[z1],
         [z2],
         [z3]])

```

```python
R = np.dot(np.linalg.inv(M), z)
print(R)
```

```python
def f(x):
    return R[0][0] * x**2 + R[1][0] * x + R[2][0]
```

```python
theorical = [f(x) for x in dominus]
plt.plot(dominus, functus, 'o', label='data')
plt.plot(dominus, theorical, label='theorical')
plt.legend()
plt.show()
```
